```python
for i in range(1,20):
    URL=f'https://diwanegypt.com/product-category/books/english-adults/page/{i}/?orderby=popularity'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'
    }
    response=requests.get(URL,headers=headers)
    print(response)
    soup=BeautifulSoup(response.content,'html.parser')
    title_tag=soup.find_all('h2',{'class':'woocommerce-loop-product__title'})
    Author_tag=soup.find_all('span',{'class':'author'})
    Price_tag=soup.find_all('span',{'class':'price'})
    link_tag=soup.find_all('a',{'class':'woocommerce-LoopProduct-link woocommerce-loop-product__link'})
    for i in range(len(title_tag)):
        Title.append(title_tag[i].text.strip())
        Author.append(Author_tag[i].text.strip())
        Price.append(Price_tag[i].text.strip())
        links.append(link_tag[i]['href'])
```

```python
Category=[]
Description=[]
for link in links:
    req_n = requests.get(link)
    print(req_n)
    html_text_n = req_n.content
    soup_n = BeautifulSoup(html_text_n, 'html.parser')
    Category_tag=soup_n.find_all('span',{'class':"posted_in"})
    DESC_tag=soup_n.find_all('p')
    for i in range(len(Category_tag)):
        Category.append(Category_tag[i].text.strip())
```

```python
MY_DIC={'BookTitle':Title,'Author':Author,'PRICE':Price,'URL':links,'Category':Category}
df=pd.DataFrame(MY_DIC)
df
```

## To add a rate column based on the order of the data, divided into 4 equal quartiles (25% per quartile)

```python
num_rows = len(df)
quartile_size = num_rows // 4
quartile_size
```

```python
def assign_rate(row):
    row_index = row.name  
    if row_index < quartile_size:
        return 4  
    elif row_index < 2 * quartile_size:
        return 3  
    elif row_index < 3 * quartile_size:
        return 2 
    else:
        return 1
```

```python
df['Rate'] = df.apply(assign_rate, axis=1)
rate_mapping = {
    4: 'A',
    3: 'B',
    2: 'C',
    1: 'D'
}
df['Rate']=df['Rate'].map(rate_mapping)
df
```
